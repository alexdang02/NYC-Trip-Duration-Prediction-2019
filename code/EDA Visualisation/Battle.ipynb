```python
import os
import pandas as pd
import matplotlib.pyplot as plt

root_folder = "/home/trungdc/unimelb/MAST30024/asm/mast30034_2021_s2_project_1-alexdang02-1"
data_dir = os.path.join(root_folder, "Data")
SQLOutput_dir = os.path.join(root_folder, "code/SparkSQL_Output")
plot_dir = os.path.join(root_folder, "Plots")
resource_save = os.path.join(root_folder,"code", "EDA Visualisation", "stat", "General")
```

```python
fhv2018 = pd.read_pickle(os.path.join(resource_save, "fhv2018.pkl"))
fhv2019 = pd.read_pickle(os.path.join(resource_save, "fhv2019.pkl"))
fhv2020 = pd.read_pickle(os.path.join(resource_save, "fhv2020.pkl"))
fhv2018 = fhv2018/1000000
fhv2019 = fhv2019/1000000
fhv2020 = fhv2020/1000000
fhv = list(fhv2018[0].values) + list(fhv2019[0].values) + list(fhv2020[0].values)
```

```python
yellow2018 = pd.read_pickle(os.path.join(resource_save, "yellow2018.pkl"))
yellow2019 = pd.read_pickle(os.path.join(resource_save, "yellow2019.pkl"))
yellow2020 = pd.read_pickle(os.path.join(resource_save, "yellow2020.pkl"))
yellow2018 = yellow2018/1000000
yellow2019 = yellow2019/1000000
yellow2020 = yellow2020/1000000
yellow = list(yellow2018[0].values) + list(yellow2019[0].values) + list(yellow2020[0].values)
green = pd.read_pickle(os.path.join(resource_save, "green.pkl"))
green = green/1000000
green = list(green[0].values)
```

```python
labels = []
for year in [2018, 2019,2020]:
    for month in range(1,13,2):
        labels.append(f"{month}/{year}")
```

```python
from matplotlib.pyplot import figure
import numpy as np

figure(figsize=(15, 15))
fig, ax = plt.subplots()
ax.plot(fhv, label = "FVFHV", marker = "o")
ax.plot(yellow, label = "yellow")
ax.plot(green, label = "green")
# ax.set_xticklables(, labels)
ax.set_xlabel("Month", size = 15)
ax.set_ylabel("Number of trip (million)",  size = 15)
plt.setp(ax, xticks=range(0,36,2), xticklabels=labels,
        )
ax.set_xticklabels(labels, rotation=40)
fig.set_figheight(6)
fig.set_figwidth(15)
ax.legend(loc='upper right', prop={'size': 14})
ax.set_title("Trends of NYC transportation in 2018-2020 period", size = 18)
ax.vlines(range(0,36,4), 0, 25, linestyles = "dashed")
fig.savefig(os.path.join(plot_dir, "Trends.png"))
```
